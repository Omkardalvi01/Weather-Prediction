```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('weather_australia.csv')
data.head(10)
```

```python
data.info()
```

```python
data['MinTemp'].interpolate(method='linear',inplace=True)
data['MaxTemp'].interpolate(method='linear',inplace=True)
data['Rainfall'] = data['Rainfall'].fillna(data['Rainfall'].mean)
```

```python
data['Cloud3pm'].fillna(-1,inplace = True)
data['Cloud9am'].fillna(-1, inplace=True)
```

```python
data['Temp3pm'].interpolate(method='linear', inplace=True)
data['Temp9am'].interpolate(method='linear',inplace=True)
```

```python
data.dropna(subset=['RainToday'], inplace=True)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.corr(numeric_only=True)
```

```python
data.drop(columns=['Evaporation','Sunshine'], axis=1, inplace=True)
```

```python
data['WindGustSpeed'].fillna(data['WindGustSpeed'].mean(), inplace=True)
data['WindGustDir'].fillna(data['WindGustDir'].mode()[0], inplace=True)
data['WindDir9am'].fillna(data['WindDir9am'].mode()[0], inplace=True)
data['WindDir3pm'].fillna(data['WindDir3pm'].mode()[0], inplace=True)
```

```python
data['WindSpeed9am'].interpolate(method='linear',inplace=True)
data['WindSpeed3pm'].interpolate(method='linear',inplace=True)
data['Humidity9am'].interpolate(method='linear', inplace=True)
data['Humidity3pm'].interpolate(method='linear', inplace=True)
```

```python
data['Pressure3pm'].interpolate(method='linear',inplace=True)
data['Pressure9am'].interpolate(method='linear', inplace=True)
```

```python
data.info()
```

```python
data.columns
```

```python
data['Rainfall'] = pd.to_numeric(data['Rainfall'])
```

```python
data.drop(columns=['Date'],inplace=True)
```

```python
X = data.drop(['RainTomorrow','Location'],axis=1)
Y = data['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8)
```

```python
Y_train
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
Y_train= lbl_encoder.fit_transform(Y_train)
Y_test = lbl_encoder.transform(Y_test)
```

```python
Y_train = pd.Series(Y_train)
```

```python
pd.concat([X_train.reset_index(drop=True),Y_train.reset_index(drop=True)],axis=1).corr(numeric_only=True)
```

```python
cat_columns = []
for column in X.columns:
    if data[column].dtype == 'object':
        cat_columns.append(column)
cat_columns
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder(sparse_output=False)
one_hot_train_df = one_hot.fit_transform(X_train[cat_columns])
one_hot_train = pd.DataFrame(one_hot_train_df, columns=one_hot.get_feature_names_out(cat_columns))
X_train= pd.concat([X_train.reset_index(drop=True), one_hot_train.reset_index(drop=True)],axis=1)
X_train.drop(columns=cat_columns,inplace=True)
```

```python
one_hot_test_df = one_hot.transform(X_test[cat_columns])
one_hot_test = pd.DataFrame(one_hot_test_df, columns=one_hot.get_feature_names_out(cat_columns))
X_test= pd.concat([X_test.reset_index(drop=True), one_hot_test.reset_index(drop=True)],axis=1)
X_test.drop(columns=cat_columns,inplace=True)
```

```python
X_train.isnull().sum()
```

```python
X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
Scaler = StandardScaler()
X_train_norm = Scaler.fit_transform(X_train)
X_test_norm = Scaler.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
prpl_comp_x = pca.fit_transform(X_train_norm)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('num_of_comp')
plt.ylabel('variance_explained')
plt.title('Scree plot')
plt.show()
```

```python
from sklearn.linear_model import  LogisticRegression
model = LogisticRegression(penalty='l2')
model.fit(X_train_norm,Y_train)
prediction = model.predict(X_test_norm)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(prediction, Y_test)
print(f'Confusion Matrix:\n {cm}')
print(classification_report(Y_test, prediction))
```
